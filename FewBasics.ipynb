ADDING SINE WAVES

```python
import matplotlib.pyplot as plt
import numpy as np

t=np.arange(0,10,0.1)

a1=np.sin(t)
a2=np.sin(2*t)
a3=np.sin(3*t)
a4=a1+a2+a3

plt.subplot(4,1,1)
plt.plot(a1)
plt.title('sine wave 1')
plt.subplot(4,1,2)
plt.plot(a2)
plt.title('sine wave 2')
plt.subplot(4,1,3)
plt.plot(a3)
plt.title('sine wave 3')
plt.subplot(4,1,4)
plt.plot(a4)
plt.title('sum of sine waves')

plt.tight_layout()
plt.show()
```

CONVOLUTION

```python
import matplotlib.pyplot as plt
import numpy as np

x=(1,2,3,4,5)
h=(2,4,6,8,10)

c=len(x)+len(h)
print(c)

x=np.pad(x,(0,c-len(x)))
h=np.pad(h,(0,c-len(h)))
print(x)
print(h)

y=[0]*c

for n in range(0,c):
    for k in range(0,c):
        y[n]=y[n]+h[k]*x[n-k]

print(y)

plt.figure(figsize=(6,8))

plt.subplot(6,1,1)
plt.plot(x)
plt.title('Continuous')
plt.subplot(6,1,2)
plt.plot(h)
plt.subplot(6,1,3)
plt.plot(y)

plt.subplot(6,1,4)
plt.stem(x)
plt.title('Discrete')
plt.subplot(6,1,5)
plt.stem(h)
plt.subplot(6,1,6)
plt.stem(y)

plt.tight_layout()
plt.show()
```

CROSS CORRELATION

```python
import matplotlib.pyplot as plt
import numpy as np

# Define two sequences
x = np.array([1, 2, 3, 4, 5])
h = np.array([2, 4, 6, 8, 10])

# Lengths of the sequences
N = len(x)
M = len(h)
L = N + M - 1  # Length of the resulting sequence after correlation

# Zero-pad sequences to match the length for correlation
x_correlation = np.pad(x, (0, L - N))
h_correlation = np.pad(h, (0, L - M))  # Pad 'h' without reversing for correlation

# Compute correlation
y_correlation = np.zeros(L)
for n in range(L):
    for k in range(N):
        if n - k >= 0 and n + k < M:
            y_correlation[n] = y_correlation[n]+ x_correlation[k] * h_correlation[n + k]

# Plotting
plt.figure(figsize=(8, 6))

plt.subplot(3, 1, 1)
plt.stem(x_correlation)
plt.title('Sequence x[n]')

plt.subplot(3, 1, 2)
plt.stem(h_correlation)
plt.title('Sequence h[n]')

plt.subplot(3, 1, 3)
plt.stem(y_correlation)
plt.title('Cross-Correlation Result')

plt.tight_layout()
plt.show()
```

AUTO CORRELATION

```python
import matplotlib.pyplot as plt
import numpy as np

# Define a sequence
x = np.array([1, 2, 3, 4, 5])

# Length of the sequence
N = len(x)
L = 2 * N - 1  # Length of the resulting autocorrelation sequence

# Zero-pad the sequence for autocorrelation
x_autocorr = np.pad(x, (0, N - 1))

# Compute autocorrelation
y_autocorr = np.zeros(L)
for n in range(L):
    for k in range(N):
        if n - k >= 0 and n + k < N:
            y_autocorr[n] += x_autocorr[k] * x_autocorr[n + k]

# Plotting
plt.figure(figsize=(8, 6))

plt.subplot(3, 1, 1)
plt.stem(x_autocorr)
plt.title('Sequence x[n]')

plt.subplot(3, 1, 2)
plt.stem(x_autocorr)
plt.title('Sequence x[n]')

plt.subplot(3, 1, 3)
plt.stem(y_autocorr)
plt.title('Autocorrelation Result')

plt.tight_layout()
plt.show()
```
