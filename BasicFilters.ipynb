HIGH PASS FILTER

```python
import numpy as np
import matplotlib.pyplot as pt
import scipy.signal as sc

fs = 3000
t = np.arange(0, 0.05, (1/fs))
s1 = np.sin(2*np.pi*50*t) #low frequency noise
s2 = np.sin(2*np.pi*500*t) #high frequency signal
sum1 = s1+s2  #noisy signal

fc = 250
b, a = sc.butter(2, fc/(fs/2), 'highpass')
y = sc.filtfilt(b, a, sum1)

pt.subplot(4, 1, 1)
pt.plot(t,s1)
pt.title('low frequency noise')
pt.subplot(4, 1, 2)
pt.plot(t,s2)
pt.title('high frequency wave')
pt.subplot(4, 1, 3)
pt.plot(t,sum1)
pt.title('noisy signal')
pt.subplot(4, 1, 4)
pt.plot(t,y)
pt.title('filtered output')
pt.tight_layout()
pt.show()
```

BAND PASS FILTER

```python
import numpy as np
import matplotlib.pyplot as pt
import scipy.signal as sc

fs = 3000
t = np.arange(0, 0.05, (1/fs))
s1 = np.sin(2*np.pi*50*t) #low frequency noise
s2 = np.sin(2*np.pi*500*t) #high frequency noise
s3 = np.sin(2*np.pi*300*t) #signal
sum1 = s1+s2+s3  #noisy signal

fc = 250
b, a = sc.butter(4, [(200*2/fs),(400*2/fs)],'bandpass')
y = sc.filtfilt(b, a, sum1)

pt.subplot(5, 1, 1)
pt.plot(t,s1)
pt.title('low frequency noise')
pt.subplot(5, 1, 2)
pt.plot(t,s2)
pt.title('high frequency noise')
pt.subplot(5, 1, 3)
pt.plot(t,s3)
pt.title('signal')
pt.subplot(5, 1, 4)
pt.plot(t,sum1)
pt.title('noisy signal')
pt.subplot(5, 1, 5)
pt.plot(t,y)
pt.title('filtered output')
pt.tight_layout()
pt.show()
```
